```python
import pandas as pd
import numpy as np
```

```python
a = {
    'A':['A0', 'A1', 'A2', 'A3'],
    'B':['B0', 'B1', 'B2', 'B3'],
    'C':['C0', 'C1', 'C2', 'C3'],
    'D':['D0', 'D1', 'D2', 'D3']
    }
b = {
    'A':['A5', 'A6', 'A7', 'A8'],
    'B':['B5', 'B6', 'B7', 'B8'],
    'C':['C5', 'C6', 'C7', 'C8'],
    'D':['D5', 'D6', 'D7', 'D8']
}

df1 = pd.DataFrame(a)

df2 = pd.DataFrame(b)
```

```python
# Concat Axix 0 (deafualt)
pd.concat([df1, df2], ignore_index=True)
```

```python
# Concat Axix 1
pd.concat([df1, df2], axis=1)
```

## Merge

```python
l = pd.DataFrame({
    'key':['k0', 'k1', 'k2', 'k3'],
    'B':['B0', 'B1', 'B2', 'B3'],
    'C':['C0', 'C1', 'C2', 'C3'],
    })

r = pd.DataFrame({
    'key':['k0', 'k1', 'k2', 'k3'],
    'D':['D0', 'D1', 'D2', 'D3'],
    'E':['E0', 'E1', 'E2', 'E3'],
    })
```

```python
l
```

```python
r
```

```python
l.merge(r, on='key')
```

```python
il = pd.DataFrame({
    'key':['k0', 'k1', 'k2', 'k3'],
    'B':['B0', 'B1', 'B2', 'B3'],
    'C':['C0', 'C1', 'C2', 'C3'],
    })

r = pd.DataFrame({
    'key2':['k0', 'k1', 'k2', 'k3'],
    'D':['D0', 'D1', 'D2', 'D3'],
    'E':['E0', 'E1', 'E2', 'E3'],
    })
```

```python
l.merge(r, left_on='key', right_on='key2')
```
