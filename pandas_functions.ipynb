```python
title_column = df_titanic.iloc[:0,:]
e = [i for i in title_column]
e

a = [i for i in e if i != 'zero']

b = []

for i in e:
    for j in range(len(e)):
        if i == f'zero.{j}':
            b.append(i)

b

# a = [[i for j in range(len(e)) if i == f'zero.{j}'] for i in e]
# a

df_tic_u = df_titanic.drop(b, axis=1)
df_tic_u = df_tic_u.drop(['zero'], axis=1)

df_tic_u.describe()
```

### Filtramos nuestro DataFrame para mayores de 60 años, ordenado de manera descendente y finalmente convertimo a una tabla en HTML

```python
df_tic_u[df_tic_u['Age']>60].sort_values('Age').to_html('./data/old_pass.html')
```
