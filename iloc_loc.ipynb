# Uso de Iloc y Loc

```python
import pandas as pd
```

```python
df_books = pd.read_csv('data/bestsellers.csv', sep=',', header=0)
df_books
```

```python
# Hacemos un slicing para ver los primeros 8 registros pares
df_books[0:9:2]
```

```python
# Vemoos la columna names y author
df_books[['Name', 'Author']]
```

### Uso de LOC

```python
# Usamos loc para hacer un slicing

df_books.loc[0:4]
```

```python
# Hacesmos un slicing sin loc y nos devuelve una serie del estar al end-1
df_books[0:4]
```

```python
# Hacemos un slicing con loc con Nombre y Author y nos devuelve una serie
df_books.loc[0:4, ['Name', 'Author']]
```

```python
# Esto es para hacer una columna de 0s
df_books.loc[:, 'Reviews'] * 0 
```

```python
# Devuelve una serie con los valores True o False
df_books.loc[:, ['Author']] == 'JJ Smith'
```

### Uso de ILOC

```python
df_books.iloc[:]
```

```python
# El sugundo argumento es el indice de la fila
df_books.iloc[:2, :4]
```

```python
# Esto es para mostrar el datal de la fila 1 y columna 3
df_books.iloc[1,3] 
```

```python
# Esto muestra el dataframe de la final 1 al 3 con la columna 1 al 3
df_books.iloc[1:3, 1:3]
```

```python
# Esto es para hacer una fila y una columna
df_books.iloc[:3, 3:]
```

- Loc : Por labels
- Iloc : por indice
